```python
import cv2

ImageFile = "./data/Lenna.png"
IMG = cv2.imread(ImageFile) # 컬로로 이미지를 출력한다.
print(IMG.shape) #행렬로 가로 크기와 세로 크기를 출력하며 컬러라 3개의 값이 나온다.

ImageFile = "./data/Lenna.png"
IMG2 = cv2.imread(ImageFile, 0) # 흑백으로 이미지를 출력한다.
print(IMG2.shape) 

cv2.imshow("Lenna Color",IMG)
cv2.imshow("Lenna GrayColor",IMG2)

cv2.waitKey() # 아무키나 입력 받으면 destroy가 된다
cv2.destroyAllWindows() #이 두개의 함수가 존재해야지 imshow가 실행이 된다.
```

```python
#수정한 이미지를 저장하는 방법

cv2.imwrite("./Output/Lenna.bmp",IMG) #파일을 다음 경로로 저장을 한다.
cv2.imwrite("./Output/Lenna.png",IMG)
cv2.imwrite("./Output/Lenna2.png",IMG, [cv2.IMWRITE_PNG_COMPRESSION, 4]) #이건 PNG의 압축률을 0~9 중 가장 높은 9로 압축해라
cv2.imwrite("./Output/Lenna2.jpg",IMG, [cv2.IMWRITE_JPEG_QUALITY, 95]) #이건 품질을 95로 해줘라를 의미한다.
```

```python
#컬러 이미지 출력 방법
from matplotlib import pyplot as plt

plt.axis("off") #하나의 축을 의미한다. 있어도 되고 없어도 되고

#opencv에서는 BGR로 이미지를 처리하기 때문에
#BGR을 RGB러 비그ㅏ즈ㅏ애힌디.
RGBIMG = cv2.cvtColor(IMG, cv2.COLOR_BGR2RGB)
plt.imshow(RGBIMG) #plt에 있는 imshow라서 opencv의 imshow랑 다르다.
plt.show()

#opencv imshow는 새로운 윈도를 만들어서 출력하지만
#plt의 imshow는 이렇게 화면에 출력을 해준다.
```

```python
 #흑백으로 이미지 출력하기
GrayIMG = cv2.imread(ImageFile, cv2.IMREAD_GRAYSCALE)

plt.axis("off")
#cmpa는 컬러 맵을 회색으로 하고 interploation은 저렇게 해라,,,
#`interpolation='bicubic'` 부분은 그림을 확대하거나 축소할 때 화면에 자연스럽게 표현되도록 하는 옵션입니다.
plt.imshow(GrayIMG, cmap="gray" , interpolation='bicubic')
plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.subplots_adjust(left = 0, bottom = 0, right = 0.5, top = 0.5)
plt.imshow(GrayIMG, cmap="gray" , interpolation='bicubic')
plt.show()
```

```python

fig, ax = plt.subplots(2,2, figsize=(5,5), sharey=True)
ax[0][0].axis("off")
ax[0][0].imshow(RGBIMG, aspect = "auto")
ax[0][1].axis("off")
ax[0][1].imshow(GrayIMG, cmap="gray" , interpolation='bicubic')
ax[1][0].axis("off")
ax[1][0].imshow(GrayIMG, cmap="gray" , interpolation='bicubic')
ax[1][1].axis("off")
ax[1][1].imshow(RGBIMG, aspect = "auto")

plt.subplots_adjust(left = 0, bottom = 0, right = 1, top = 1, wspace=0.05, hspace=0.05)
plt.show()
```
