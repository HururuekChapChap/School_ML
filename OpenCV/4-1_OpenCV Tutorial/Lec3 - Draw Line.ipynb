```python
import cv2
import numpy as np

#0으로 채워져 있는 크기는 512,512, 그리고 컬러 영상을 만들어라
#그리고 타입은 8비트 짜리 데이터 타입을 만들어라
#값이 0인 영상은 색깔이 검정색인 영상을 만들어라 
#흰색 바탕을 하고 싶을 경우에는 검정(0)에 255를 더하면 된다.
img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255

#두 점을 찍어 준거다
pt1 = 100, 100
pt2 = 400, 400
#처음에는 좌상 상단, 오른쪽 색깔은 초록색, 두깨는 2
cv2.rectangle(img,pt1,pt2,(0,255,0),2)

#라인도 두개 그린다 각각 푸른색 빨강색 두깨는 5
cv2.line(img,(0,0), (500,0), (255,0,0), 5)
cv2.line(img,(0,0), (0,500), (0,0,255),5)

#img 라는 이름으로 img를 출력해라
cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 직선과 교점이 발생하는 사각형 그리기

img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255

x1, x2 = 100, 400
y1, y2= 100, 400

#빨간색 사격형을 그린다.
cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,255))

pt1 = 120, 50
pt2 = 300, 500
#선을 그린다. 두께는 2
cv2.line(img, pt1,pt2,(255,0,0),2)

#4개의 값을 집어 넣는다. left top right bottom 너비와 높이를 지정해준다.
imgRect = (x1,y1, x2-x1, y2-y1)
#결국 사격형이 imRect가 된다. 그리고 rp1과 rpt2가 교점이 된다.
retval, rpt1, rpt2 = cv2.clipLine(imgRect,pt1,pt2)

if retval:
    #원을 그려라를 의미한다. -1이면 원을 채워라 반지름이 5이고 원을 채워서 그려라
    cv2.circle(img, rpt1, radius=5, color=(0,255,0), thickness= -1)
    cv2.circle(img, rpt2, radius=5, color=(0,255,0), thickness= -1)
    
cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#원 그리기

img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255
#중점을 구산다.
cx = img.shape[0] // 2
cy = img.shape[1] // 2
#200에서 0 까지 -100 씩 줄인다.
for r in range(200,0,-100):
    #센터랑 반지름을 가져온다 200에서 0으로 100씩 줄이니 2번 그려진다.
    cv2.circle(img,(cx,cy), r, color=(255,0,0))

    #원안에 채워서 그려라!
cv2.circle(img,(cx,cy), radius = 50, color=(0,0,255), thickness = -1)

cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#타원형 그리기

img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255
#중점을 구산다.
cx = img.shape[0] // 2
cy = img.shape[1] // 2

print(img.shape[0]) # X좌표
print(img.shape[1]) # Y좌표
print(img.shape[2]) # 컬러라서 3차원

ptCenter = cx , cy
size = 200, 100

#0과 45는 기울어진 정도를 의미하고
#0,360은 타원 전체를 의미한다. 0,180은 타원의 절반(부채꼴 처럼 진짜 절반만)만 그리겠다.
cv2.ellipse(img,ptCenter,size, 0,0,180,(255,0,0))
cv2.ellipse(img,ptCenter,size, 45,0,360,(0,0,255))

#박스의 센터랑 박스의 사이즈 
box = (ptCenter, size, 0)
cv2.ellipse(img,box,(255,0,0), 5)

box = (ptCenter, size, 45)
cv2.ellipse(img,box,(0,0,255), 5)
       
cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#다각형 그리기

img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255
#점들의 집합을 그린다. 점이 4개이니 직사각형, 점이 세개면 삼각형
pts1 = np.array([[100,100],[200,100],[200,200],[100,200]])
pts2 = np.array([[300,200],[400,100],[400,200]])

#그려봐라 다각형을
cv2.polylines(img,[pts1,pts2], isClosed=True, color=(255,0,0))
                 
cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()  

```

```python
#타원형 안에 다각형 그리기
img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255

cx = img.shape[0] // 2
cy = img.shape[1] // 2

ptCenter = cx , cy
size = 200, 100
#타원형을 그리고
cv2.ellipse(img,ptCenter,size, 0,0,360,(255,0,0))
#타원형 안에 다각형을 그린다.
pts1 = cv2.ellipse2Poly(ptCenter, size, 0,0,360, delta=30)

cv2.ellipse(img,ptCenter,size, 45,0,360,(255,0,0))
pts2 = cv2.ellipse2Poly(ptCenter, size, 45,0,360, delta=45)
#배열에 들어가 있는 것들을 그려준다.
cv2.polylines(img, [pts1,pts2], isClosed=True, color=(0,0,255))

cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#회전하는 원그리기
img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255

x,y =256, 256
size = 200

for angle in range(0,90,10):
    #윗쪽과 아래쪽으 점을 찍어주고 size의 너비와 높이를 해준다음, angle은 각도를 의미한다.
    rect = ((256,256), (size, size), angle)
    #rect는 회전된 정보이다, 각 사각형의 모서리 정보를 반환해주는 함수이다.
    box = cv2.boxPoints(rect).astype(np.int32)
    r = np.random.randint(256)
    g = np.random.randint(256)
    b = np.random.randint(256)
    
    cv2.polylines(img,[box], True, (r,g,b),2)
    
cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#색을 채우는 방법
img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255

pts1 = np.array([[100,100],[200,100],[200,200],[100,200]])
pts2 = np.array([[300,200],[400,100],[400,200]])

cv2.fillConvexPoly(img,pts1, color=(255,0,0))
cv2.fillPoly(img,[pts2], color=(0,0,255))

cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#문자열

img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255

text = "openCV Programming"
org = (50,100)
font = cv2.FONT_HERSHEY_SIMPLEX
#이미지에 org 점에서 폰트를 넣어준다.
cv2.putText(img,text,org,font,1,(255,0,0),2)

#size는 글자의 크기를 받아온다.
size, basLine = cv2.getTextSize(text,font,1,2)
#두점을 찍어준다.
cv2.rectangle(img,org,(org[0]+size[0],org[1]-size[1]), (0,0,255))
cv2.circle(img,org,3,(0,255,0),2)

cv2.imshow("img", img)
cv2.waitKey()
cv2.destroyAllWindows()
```
