```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
src = np.array(
    [
        [0,0,0,0],
        [1,1,3,5],
        [6,1,1,3],
        [4,3,1,7]
    ], dtype=np.uint8
)
#Lmax가 dtype = uint8이기 때문에 255이다. 따라서 0-7 사이의 수가 나오는게 아니라 0-255 숫자가 나오게 된다.
#따라서 이 영상은 0 - 255 까지 중에 0 - 7 까지만 있는 매우 몰려 있는 영상이다.
dst = cv2.equalizeHist(src)
print("dst = ", dst)
```

```python
src = cv2.imread("./data/Lenna.png", 0)
dst = cv2.equalizeHist(src)
cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.title("Grayscale histogram of Lenna.png")

hist1 = cv2.calcHist( images=[src], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist1, color="b", label="hist1 in src")
hist2 = cv2.calcHist( images=[dst], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist2, color="r", alpha=0.7, label="hist2 in dst")
plt.legend(loc="best")
plt.show()

# 이것을 본다면 파랑색이 원래 영상인데 0 과 255 지점에는 영상이 없지만 
# 이것을 histequlization을 했을 때 빨강색이 넓게 분포가 된 것을 확인 할 수 있다.
```

```python
src = cv2.imread("./data/Lenna.png")
cv2.imshow("src", src)

#1 hsv로 바꿔서 사용
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV) #hsv로 변경한다. H:색상 S:채도 V:명도
h,s,v = cv2.split(hsv)
v2 = cv2.equalizeHist(v) #명도 값만 가지고 equlizatio을 한다 색상 값을 했을 때 색상이 변해버린다.
hsv2 = cv2.merge([h,s,v2])
dst = cv2.cvtColor(hsv2, cv2.COLOR_HSV2BGR)
cv2.imshow("HSV - V", dst)


#1 - 1 hsv로 바꿔서 사용(색상변경)
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV) #hsv로 변경한다. H:색상 S:채도 V:명도
h,s,v = cv2.split(hsv)
h2 = cv2.equalizeHist(h) #명도 값만 가지고 equlizatio을 한다 색상 값을 했을 때 색상이 변해버린다.
hsv2 = cv2.merge([h2,s,v])
dst = cv2.cvtColor(hsv2, cv2.COLOR_HSV2BGR)
cv2.imshow("HSV - H", dst)

#1 - 2hsv로 바꿔서 사용(채도변경)
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV) #hsv로 변경한다. H:색상 S:채도 V:명도
h,s,v = cv2.split(hsv)
s2 = cv2.equalizeHist(s) #명도 값만 가지고 equlizatio을 한다 색상 값을 했을 때 색상이 변해버린다.
hsv2 = cv2.merge([h,s2,v])
dst = cv2.cvtColor(hsv2, cv2.COLOR_HSV2BGR)
cv2.imshow("HSV - S", dst)

#2
yCrCv = cv2.cvtColor(src, cv2.COLOR_BGR2YCrCb)#YCrCv에서 Y는 밝기 정보
y, Cr, Cv = cv2.split(yCrCv)
y2 = cv2.equalizeHist(y)
yCrCv2 = cv2.merge([y2, Cr, Cv])
dst2 = cv2.cvtColor(yCrCv2, cv2.COLOR_YCrCb2BGR)
cv2.imshow("yCrCv", dst2)
cv2.waitKey()
cv2.destroyAllWindows()

# 결과가 차이가 발생한다. 
# HSV는 약간 색깔이 찐해졌어요. - HSV는 색상이 진해지는 경향이 있다.
# 그리고 ycrcv는 밝아졌어요 밝으면서 살짝 찐해진 경우 - 색채에 대한 정보가 거의 없다.
```
