```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
src = np.array(
[
    [0,0,0,0],
    [1,1,3,5],
    [6,1,1,3],
    [4,3,1,7]
    ], dtype = np.uint8
        )
#소스를 영상으로 하지 않고 4*4 행렬을 정의합니다

# 히스토리그램을 구합니다
# channels = 0이라는 것은 채널을 하나만 쓴다
# 히스트 사이즈는 빈 사이즈를 의미하고 여기서는 빈을 4개 쓰고
# ranges = 0,8은 0 - 7 까지 범위를 정한다는 것을 의미한다.
hist1 = cv2.calcHist(
    images=[src], channels=[0], mask=None, histSize=[8], ranges=[0,8]
)
print("his1 = ", hist1)
# 여기에서는 범위가 0 부터 7 까지인데 빈의 갯수가 4개이므로 한 배열에 두개가 들어간다
hist2 = cv2.calcHist(
images=[src], channels=[0], mask = None, histSize=[4], ranges=[0,8]
)
print("hist2=",hist2)
```

```python
src = cv2.imread("./data/Lenna.png", 0)

# 압축한것 1/8
hist1 = cv2.calcHist(images=[src], channels=[0], mask=None, histSize=[32], ranges=[0,256])
#원래의 이미지 값
hist2 = cv2.calcHist(images=[src], channels=[0], mask=None, histSize=[256], ranges=[0,256])

#1 flatten은 일차원 리스트로 만들어 주는 것  - 주로 딥러닝에서 사용한다
hist1 = hist1.flatten()
hist2 = hist2.flatten()

# print(hist1)
# print(hist2)

#2 flatten을 해주고 찍어주는 것이다.
plt.title("hist1: binX = np.arange(32)")
plt.plot(hist1,color="r")
# 빈에 32개를 넣어주는 것이다
binX = np.arange(32)
# 막대기를 그려주는 것이다
plt.bar(binX, hist1, width=1, color="b")
plt.show()

#3
plt.title("hist1: binX = np.arange(32)*8")
binX = np.arange(32) *8
plt.plot(binX,hist1,color="r")
plt.bar(binX, hist1, width=8, color="b")
plt.show()

#4
plt.title("hist2: binX = np.arange(256)")
plt.plot(hist2,color="r")
binX = np.arange(256)
plt.bar(binX, hist2, width=1, color="b")
plt.show()
```

```python
src = cv2.imread("./data/Lenna.png")
histColor = ("b", "g","r")
for i in range(3):
    hist = cv2.calcHist(
        images=[src], channels=[i], mask=None, histSize=[256],ranges=[0,256]
    )
    plt.plot(hist,color = histColor[i])
plt.show()
```
