```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("./data/Lenna.png")

rows, cows, channel = img.shape

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

H, S, V = cv2.split(hsv)

changeV = V - 100

changeV[100:200 , 100:200] = 255

merge = cv2.merge([H,S,changeV])
RGBIMG = cv2.cvtColor(merge, cv2.COLOR_BGR2RGB) # BGR RGB  .
#cv2.imshow("HSV", hsv)
#cv2.imshow("V",V)
#cv2.imshow("Merge", merge)
#cv2.waitKey()
#cv2.destroyAllWindows()
plt.imshow(RGBIMG)
plt.show()
```
