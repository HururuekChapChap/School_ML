```python
import cv2
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
src1_bgr = cv2.imread("./data/book1.jpg")
src2_bgr = cv2.imread("./data/book2.jpg")

img1 = cv2.cvtColor(src1_bgr , cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(src2_bgr , cv2.COLOR_BGR2GRAY)
src1_rgb = cv2.cvtColor(src1_bgr , cv2.COLOR_BGR2RGB) #plt로 보여주기 위해서는 bgr -> rgb로 바꿔줘야한다.
src2_rgb = cv2.cvtColor(src2_bgr , cv2.COLOR_BGR2RGB)

plt.figure(figsize = (10,30))
image = plt.subplot(1,2,1)
image.set_title("img1")
plt.axis("off")
plt.imshow(src1_rgb)

image = plt.subplot(1,2,2)
image.set_title("img2")
plt.axis("off")
plt.imshow(src2_rgb)
```

```python
#1 descriptor (orb , brisk)

orb = cv2.ORB_create(nfeatures = 1000)
brisk = cv2.BRISK_create()
detector = [orb, brisk] #배열 안에 두개의 객체를 넣음

didx = 0
#detect로 KeyPonit로 찾고 Compute로 descriptor 정보가 나온다
kp1, des1 = detector[didx].detectAndCompute(img1, None)
kp2, des2 = detector[didx].detectAndCompute(img2, None)

# 2. matcher (bf , flann)

#Hamming Distance 는 머신러닝 수업 때 말했는데,
# 서로 다른 비트가 몇개인지 검사하는 것이다
# 그래서 5개의 비트가 다르면 Hamming Distance 는 5이다
bf = cv2.BFMatcher_create(cv2.NORM_HAMMING, crossCheck = True)
flann = cv2.FlannBasedMatcher_create()
matcher = [bf, flann]

midx = 0

# 따라서 4가지의 경우의 수가 나온다 orb - bf , orb - flann , brisk - bf, brisk - flann
#현재는 orb - bf 형태로 설정 되어 있다
#바꾸고 싶으면 위에 didx  midx 를 바꿔저야한다.

#bf 경우에는 match에 그냥 넣어주면 되는데
#flann은 float로 변형 해서 넣어줘야한다.
try:
    matches = matcher[midx].match(des1, des2)
except:
    matches = matcher[midx].match(np.float32(des1), np.float32(des2))
    
#result display

#매칭 결과가 3가지의 값을 가지는데, 두개는 매칭된 점, distance는 거리
matches = sorted(matches, key = lambda x:x.distance)
print("len(matches)=", len(matches))
for i, m in enumerate(matches[:10]):
    # img1 KeyPoint , img2 KeyPoint, keyPonit 끼리의 거리
    print('matches[{}]=(queryldx:{}, trainldx:[{}, distance:{})'.format(i,m.queryIdx,m.trainIdx, m.distance))

    #알아서 그려서 넣어준다.
    #상위 몇개만 출력하고 싶다면 matches[:10]
img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches, None, flags=2)
plt.figure(figsize=(30,60))
plt.imshow(img3)
plt.show()

#매칭 된 것이 340개 이고 상위 10개만 출력해봐라
#첫번 째 310번째 KeyPoint가 두번째 이미지의 215 KeyPoint랑 거리가 10으로 매칭 되었다.
```
