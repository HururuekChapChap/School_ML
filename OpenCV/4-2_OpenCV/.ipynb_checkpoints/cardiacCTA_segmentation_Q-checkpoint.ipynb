```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as metrics

from skimage import filters
```

```python
def calcSim(target, predict):
    
    if len(target) != len(predict):
        return -1
    
    num_target = 0
    num_predict = 0
    num_union = 0    
    for i in range(len(target)):
        if (target[i] != 0):
            num_target = num_target + 1            
        if (predict[i] != 0):
            num_predict = num_predict + 1
        if (target[i] != 0) and (target[i] == predict[i]):
            num_union = num_union + 1
            
    sim = 2*(num_union/(num_target+num_predict))
    
    return sim
```

```python
src = cv2.imread('./data/cardiac_cta.bmp', cv2.IMREAD_GRAYSCALE)
target = cv2.imread('./data/cardiac_label.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
## TODO : Segmentation

ret2, predict = cv2.threshold(src,195, 255, cv2.THRESH_BINARY)

#쓰레쉬 홀드 적용후 CCL로 사각형 영역을 구해준다.
ret, labels, stats, centroids = cv2.connectedComponentsWithStats(predict)
# print(ret)

maxX = maxY = 0
maxWidth = maxHeight = 0
size = 0
maxPoint = 0

#가장 사각형이 큰 것을 찾아준다.
for i in range(1, ret):
    x, y, width, height, area = stats[i]
    
    if(area > size):
        maxX = x
        maxY = y
        maxWidth = width
        maxHeight = height
        size = area
        maxPoint = i
    
    #바운딩 박스를 만들어준다.
#     cv2.rectangle(predict, (x,y), (x+width, y + height), (255,255,255), 2)
    
# print(maxX , maxY , maxWidth , maxHeight)

#가장 사각형이 큰 부분을 제외하고 다 0으로 만들어 버린다.
for (y_index , y) in enumerate(predict):
    for (x_index , x) in enumerate(y):
        if not(y_index >= maxY and y_index < maxY + maxHeight and x_index >= maxX and x_index <= maxX + maxWidth):
            predict[y_index][x_index] = 0

            
#커널을 등록 해준다.
kernel = cv2.getStructuringElement(shape=cv2.MORPH_RECT, ksize=(3,3))
predict = cv2.dilate(predict, kernel, iterations = 1)
# predict = cv2.erode(predict,kernel, iterations = 1)

# 195 (3,3) erode X dilate 1 = 95.18

```

```python
plt.figure(figsize=(15, 15))
plt.subplot(1, 3, 1)
plt.axis('off')
plt.imshow(src, cmap='gray')

plt.subplot(1, 3, 2)
plt.axis('off')
plt.imshow(predict, cmap='gray')

plt.subplot(1, 3, 3)
plt.axis('off')
plt.imshow(target, cmap='gray')
```

```python
target = target.flatten()
predict = predict.flatten()
print(target.shape)
print(predict.shape)

target[target[:] == 255] = 1
predict[predict[:] == 255] = 1

sim = calcSim(target, predict)
print("similarity: ", sim)
```
