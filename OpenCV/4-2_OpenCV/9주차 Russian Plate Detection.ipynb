```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#   
imagePath = "./data/russianplate.jpg"
cascPath = "./xml/haarcascade_russian_plate_number.xml"

#Create the haar cascade
rpcascade = cv2.CascadeClassifier(cascPath)

#Read the image
image = cv2.imread(imagePath)
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Detect faces in the image
plate = rpcascade.detectMultiScale(gray,minNeighbors=5,minSize=(20,20))

#Draw bouding box
for (x,y,w,h) in plate:
    cv2.rectangle(imageRGB,(x,y),(x+w,y+h),(255,0,0),2)
    
#display
plt.figure(figsize=(15,15))
plt.imshow(imageRGB)
```
