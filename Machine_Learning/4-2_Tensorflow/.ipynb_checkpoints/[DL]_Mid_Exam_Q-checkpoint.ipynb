```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

print(tf.__version__)
```

```python
input_file = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
data = np.loadtxt(input_file, delimiter=',')
x, y = data[:, :-1], data[:, -1]
print(data.shape)
train_X, test_X, train_Y, test_Y = train_test_split(x, y, random_state=0)
print(train_X.shape)
print(train_Y.shape)
print(test_X.shape)
print(test_Y.shape)
```

```python
# TODO: 코드 구현 
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)

train_Y = tf.keras.utils.to_categorical(train_Y, num_classes=2)
test_Y = tf.keras.utils.to_categorical(test_Y, num_classes=2)

model = tf.keras.Sequential(
    [
     tf.keras.layers.Dense(units= 47, activation="relu", input_shape=(8,)),
     tf.keras.layers.Dense(units= 33, activation="relu"),
      tf.keras.layers.Dense(units= 2, activation="softmax")
    ]
)

model.compile(optimizer = tf.keras.optimizers.Adam(lr=0.07), loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()

history = model.fit(train_X, train_Y, epochs=200, batch_size=30)
```

```python
print("훈련 세트 정확도")
print(model.evaluate(train_X, train_Y, verbose=0))
print("테스트 세트 정확도")
print(model.evaluate(test_X, test_Y, verbose=0))
```
