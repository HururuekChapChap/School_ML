```python
#훈련 데이터와 테스트 데이터를 나눠야한다.

from sklearn.model_selection import train_test_split

#w자동으로 0.75 : 0.25로 나눠준다. 그러나 데이터가 모여 있기 때문에, 잘 못하면 
# 하나로 묶여서 있을 수 있기 때문에 random_state로 해준다.
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset["data"], iris_dataset["target"], random_state=0
)

#입력 = data , 출력 = target

print("X_train size : ", X_train.shape)
print("y_train size : ", y_train.shape)
# 트레인 데이터로 훈련을 하고 

print("X_test size : ", X_test.shape)
print("y_test size : ", y_test.shape)
#테스트 데이터로 출력을 하도록 한다
```

```python
# 데이터를 나누고 나서 데이터 살펴보기

import pandas as pd
from preamble import *

#열의 이름은 iris_datset, feature_names에 있는 문자열을 사용합닏.
iris_dataframe  = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
#데이터 프레임을 사용해 y_train에 따라 색으로 구분된 산점도 행렬을 만듭니다.
pd.plotting.scatter_matrix(
    iris_dataframe, c=y_train, figsize=(15,15), marker="0",
    hist_kwds={"bins" : 20}, s= 60, alpha = .8 , cmap=mglearn.cm3                      
                          )

#이 삼전도를 보는 이유는 4차원이면 화면에 표실 할 수가 업삳.
"""
 이것을 화면에 표시해서 feature 간의 관계를 보기 위해서는 삼전도 행렬을 사용해야한다.
 
"""
```

```python
# 첫번째 머신러닝 모델 : K- 최근접 이웃 알고리즘
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
"""
knn은 실제로 트레이닝을 하지 않는다.
왜 이 친구는 학습을 하지 않을까?
이유를 생각해 봅시다
"""
```

```python
knn.fit(X_train, y_train)
#트레인 데이터를 다 가지게 된다.
```

```python
#예측 및 모델 평가하기

X_new = np.array([[5,2.9,1,0.2]])
print(X_new.shape)
```

```python
prediction = knn.predict(X_new)
print(prediction)
"""실제 예측 값은 0이다. 이름은 세토사"""
print(iris_dataset["target_names"][prediction])
```

```python
y_pred = knn.predict(X_test)


print(y_pred)
print(format(np.mean(y_pred == y_test))) """y_pred는 우리가 예측한 알고리즘이고 이것이 y_test랑 같다면 정답이다"""
print(format(knn.score(X_test, y_test))) 
#둘다 예측값을 알려주는 같은 방법이다.완전히 다 같을 경우에는 1이고 완전히 다 다르면 0일 것이다.
```

```python
#요약!!!
X_train, X_test, y_trian, y_test = train_test_split(
    iris_dataset["data"], iris_dataset["target"], random_state=0
)
                 
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print(format(knn.score(X_test, y_test)))
```
