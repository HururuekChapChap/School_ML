```python
from sklearn.model_selection import train_test_split
#train_test_splt은 알아서 훈련데이터와 테스트 데이터를 나눠준다.
X, y = mglearn.datasets.make_forge()

print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
#random_state = 0은 임의적으로 섞어 준다는 것을 의미한다.
```

```python
from sklearn.neighbors import KNeighborsClassifier
#Iris에서 사용하였다,.
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
#fit은 학습하라는 의미이다.
```

```python
print("테스트 세트 예측 :", clf.predict(X_test))
print("테스트 세트 참값 :", y_test)
#predic는 예측 값을 나오게 해준다. 트레인 데이터로 예측을 해준다.
#K-nn은 딱히 학습이라는게 없다
"""
K-nn은 나눠주는 disition boundary가 필요없다 대신 데이터 자체를 셋팅 해주고 새로운 데이터가 
들어 왔을 때 가장 가까운 친구를 찾아준다.
따라서 훈련이라는게 아니다

"""
```

```python
print("테스트 세트 정확도 : {:.2f}".format(clf.score(X_test, y_test)))
#정확도만 알려면 이 함수를 사용하면 된다.
```

```python
# KNeighborsClassifier 분석
fig, axes = plt.subplots(1,3, figsize=(10,3))

for n_neighbors, ax in zip([1,3,9],axes):
#neighbor가 1일 때, 3일 때 9일때:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    
    ax.set_title("{} neighbor".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
    
axes[0].legend(loc="best")
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=66
)

training_accuracy = []
test_accuracy = []

#1 에서 10 까지  n_neighbors를 적용
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    
    training_accuracy.append(clf.score(X_train,y_train))
    
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label="traing accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
#K-Neighbors Regression
"""
 리그레션에 적용한다는 무슨 말인가? 결국 개념은 클래시피케이션과 같다 => 회귀분석이 리스레션이다.
 대신 차이점은 X값 이 주어지면 Y 값을 찾는 것이다. 이때 방법은 K 의 갯수에 따라서
 가장 가까운 K를 통해서 Y의 값을 추정하게 된다. 
"""
mglearn.plots.plot_knn_regression(n_neighbors=1)
"""
트레인 데이터가 있을 때 초록색이 테스트 데이터다이다. 즉 초록색과 가장 가까운 값을 찾아서
그 값을 그대로 취하게 된다 왜냐하면 K값이 1이기 때문이다.
"""
```

```python
mglearn.plots.plot_knn_regression(n_neighbors=3)
"""
두번째는 K를 3으로 했으며 초록색 별과 가장 가까운 점 3개를 토대로 Y값을 추정하게 된다. 
"""
```

```python
from sklearn.neighbors import KNeighborsRegressor

"""
mglearm에 데이터 웨이브라는 것이 있엇고 특징의 사이즈는 1이고 타게
"""
X, y = mglearn.datasets.make_wave(n_samples=40)
"""wave 데이터 셋을 훈련과 테스트로 나눠준다."""
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)

reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train, y_train)
"""
이렇게 하면 트레인과 훈련데이터를 잘 나눴다를 알려준다.
"""

train = X_train.shape
print(train)
test = X_test.shape
print(test)
```

```python
print("테스트 세트 예측 : \n", reg.predict(X_test))
"""
회귀분석일 경우에 정확도를 예측 할 때 알스퀘어 방식을 사용하여 예측한다.
결국 1에 가까울 수록 예측이 잘 됐음을 의미한다.
"""
```

```python
print("테스트 세트 R^2: {:.2f}".format(reg.score(X_test, y_test)))
"""
회귀분석일 경우에 정확도를 예측 할 때 알스퀘어 방식을 사용하여 예측한다.
결국 1에 가까울 수록 예측이 잘 됐음을 의미한다.
"""
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,4))

line = np.linspace(-3,3,1000).reshape(-1,1)

"""
-3 부터 3 까지 1000개 데이터의 포인터를 만들고 그것에 대한 예측 값을 구해보는 것이다.
"""

for n_neighbors, ax in zip([1,3,9], axes):
    
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, "^", c=mglearn.cm2(0),markersize=8)
    ax.plot(X_test, y_test, "v", c=mglearn.cm2(1),markersize=8)
    
    ax.set_title(
        "{}-neighbor :{:.2f} test score: {:.2f}".format(
            n_neighbors, reg.score(X_train, y_train), reg.score(X_test, y_test)))
    ax.set_xlabel("feature")
    ax.set_ylabel("target")
axes[0].legend(["model predictions", "train data/target", "test data/target"], loc ="best")

"""
 아래의 데이터가 말하는 바
 
 네이버가 1일 때 3일 때 9일때는 테스트 데이터에 대한 정보이다.
 
 K가 1일 때와 k가 3일 때 9일 때로 갈 수록 점점 왔다리갔다리가 적어진다. 즉 선이 부드러워 진다.
 
 k가 1일 때는 test score가 0.35가 나온다. 그리고 3일 때가 0.83으로 가장 높다
 
 
"""
```
