```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from preamble import *
```

```python
x, y = make_moons(n_samples=100, noise=0.25, random_state=3)
#stratify = y   y         .
xtrain, xtest, ytrain, ytest = train_test_split(x,y,stratify=y, random_state=42)
mglearn.discrete_scatter(xtrain[:,0], xtrain[:,1], ytrain)
```

```python
mlp = MLPClassifier(solver="lbfgs", random_state=0).fit(xtrain, ytrain)

mglearn.plots.plot_2d_separator(mlp, xtrain,fill=True ,alpha = 0.3)
mglearn.discrete_scatter(xtrain[:,0], xtrain[:,1], ytrain)

plt.xlabel("feature 0")
plt.ylabel("feature 1")

#  default        100  .
```

```python
mlp = MLPClassifier(solver ="lbfgs", random_state=0, hidden_layer_sizes=[10])

#  10  .

mlp.fit(xtrain, ytrain)
mglearn.plots.plot_2d_separator(mlp,xtrain,fill=True, alpha = 0.3)
mglearn.discrete_scatter(xtrain[:,0], xtrain[:,1], ytrain)
```

```python
#10    
#      10 
mlp = MLPClassifier(solver="lbfgs", random_state=0, hidden_layer_sizes=[10,10])
mlp.fit(xtrain,ytrain)
mglearn.plots.plot_2d_separator(mlp, xtrain, fill = True, alpha = 0.3)
mglearn.discrete_scatter(xtrain[:,0], xtrain[:,1], ytrain)
plt.xlabel("feature 0")
plt.ylabel("feature 1")
```

```python
#tanh    10    

#defaul . 
mlp = MLPClassifier(solver="lbfgs",activation="tanh" ,random_state=0, hidden_layer_sizes=[10,10])
mlp.fit(xtrain,ytrain)
mglearn.plots.plot_2d_separator(mlp, xtrain, fill = True, alpha = 0.3)
mglearn.discrete_scatter(xtrain[:,0], xtrain[:,1], ytrain)
plt.xlabel("feature 0")
plt.ylabel("feature 1")
```

```python
fig , axes = plt.subplots(2,4, figsize=(20, 8))
for axx, n_hidden_nodes in zip(axes, [10,100]):
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
        mip = MLPClassifier(solver="lbfgs", random_state=0, hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes], alpha = alpha)
        mip.fit(xtrain,ytrain)
        mglearn.plots.plot_2d_separator(mlp, xtrain, fill = True, alpha = 0.3, ax=ax)
        mglearn.discrete_scatter(xtrain[:,0], xtrain[:,1], ytrain, ax=ax)
        ax.set_title("n_hidden=[{},{}] \n alpha={:.4f}".format(n_hidden_nodes, n_hidden_nodes, alpha))
        
```

```python
cancer = load_breast_cancer()
print(cancer.data.shape)
print(cancer.target.shape)
print("    : \n", cancer.data.max(axis=0))
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(cancer.data, cancer.target, random_state=0)
mip = MLPClassifier(random_state=42)
mlp.fit(xtrain, ytrain)

print("  : {: 2f}".format(mlp.score(xtrain, ytrain)))
print("   : {:.2f}".format(mlp.score(xtest,ytest)))
```

```python
scaler = StandardScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)

mlp = MLPClassifier(random_state=0)
mlp.fit(xtrain_scaled, ytrain)

print("   : {: .3f}".format(mlp.score(xtrain_scaled, ytrain)))
print("   : {:.3f}".format(mlp.score(xtest_scaled, ytest)))
```

```python
mlp = MLPClassifier(max_iter=1000, random_state=0)
mlp.fit(xtrain_scaled, ytrain)

print("  : {:.3f}".format(mlp.score(xtrain_scaled, ytrain)))
print("  : {:.3f}".format(mlp.score(xtest_scaled, ytest)))
```

```python
import warnings
warnings.filterwarnings(action="ignore")

print(mlp.coefs_[0].shape)
print(mlp.coefs_[1].shape)
plt.figure(figsize=(20,5))
plt.imshow(mlp.coefs_[0])
plt.yticks(range(30), cancer.feature_names)
plt.xlabel("hidden unit")
plt.ylabel("input feature")
plt.colorbar()
```

```python
mlp = MLPClassifier(max_iter=1000, alpha = 1 ,random_state=0)
mlp.fit(xtrain_scaled, ytrain)

print("  : {:.3f}".format(mlp.score(xtrain_scaled, ytrain)))
print("  : {:.3f}".format(mlp.score(xtest_scaled, ytest)))
```

```python
import sklearn.datasets as datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
#

digits = datasets.load_digits()
x = digits.data
y = digits.target
print(x.shape)
print(y.shape)

plt.matshow(digits.images[0])
plt.show()
```

```python
xtrain , xtest, ytrain, ytest = train_test_split(x, y , stratify=y, random_state=42)
scaler = StandardScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)

mlp = MLPClassifier(max_iter=300, solver = "adam", random_state = 0, activation="relu"
                 , hidden_layer_sizes=[100,50])
mlp.fit(xtrain_scaled, ytrain)

print("   : {: .3f}".format(mlp.score(xtrain_scaled, ytrain)))
print("   : {: .3f}".format(mlp.score(xtest_scaled, ytest)))
```

```python
mnist = datasets.fetch_openml("mnist_784", version = 1)
images = mnist.data
labels = mnist.target

print(images.shape)
print(labels.shape)
```

```python
xtrain , xtest, ytrain, ytest = train_test_split(images, labels, stratify=labels, random_state=42)
scaler = StandardScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)

# verbose = True      
mlp = MLPClassifier(max_iter=50, learning_rate_init=0.01 ,solver = "sgd", random_state = 0, hidden_layer_sizes=[300, 300,300], verbose = True)
mlp.fit(xtrain_scaled, ytrain)

print("   : {: .3f}".format(mlp.score(xtrain_scaled, ytrain)))
print("   : {: .3f}".format(mlp.score(xtest_scaled, ytest)))
```
