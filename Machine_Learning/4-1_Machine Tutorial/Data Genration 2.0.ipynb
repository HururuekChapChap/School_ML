```python
from sklearn.datasets import load_breast_cancer
#사이킷 런에서 제공해주는 유방암 데이터
cancer = load_breast_cancer()
print("caner.keys() \n", cancer.keys())
print("유방암 데이터의 형태 :", cancer["data"].shape)
print("종양 종료", cancer["target_names"])
print("종양", cancer["target"])
```

```python
print("클래스별 샘플 갯수 : \n", {n:v for n, v in zip(cancer["target_names"],
                                              np.bincount(cancer["target"]))})

# 직관적으로 봤을 때 클래스 별로 샘플 갯수가 비슷하게 있어야한다. 치우쳐져 있으면 
#좋은 결과가 나올 수가 없게 된다.
```
